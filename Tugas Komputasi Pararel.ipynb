```python
import multiprocessing as mp
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
#Mendefinisikan fungsi penjumlahan vektor
def vector_addition(A,B):
  result = A+B
  time.sleep(0.1)
  return result

#Meminta input banyaknya elemen vektor
n = 100

#Menggenerasi vektor dengan input random
A = np.random.randint(-1000, 1000, n)
B = np.random.randint(-1000, 1000, n)


#Proses paralel
if __name__ == '__main__':
  running_time = []
  NumPar = [2,4,6,8,10,12,14,16]
  input= []

  for i in range(n):
    input.append((A[i],B[i]))

  for p in NumPar:
    pool = mp.Pool(p)
    t = time.time()
    result = pool.starmap(vector_addition, input)
    t2 = time.time()
    running_time.append(t2 - t)

  #plotting running time
  plt.plot(NumPar, running_time, "r-")
  plt.xlabel('Number of Cores')
  plt.ylabel('Running Time')
  plt.show()


  #Menghitung nilai speed-up beserta plottingnya
  S = []
  E = []
  for m in range(len(NumPar)):
    S.append(running_time[0]/running_time[m])

  plt.plot(NumPar, S, 'r*')
  plt.xlabel('Number of Cores')
  plt.ylabel('SpeedUp')
  plt.show()


  #Menghitung nilai efisiensi beserta plottingnya
  for l in range(len(NumPar)):
    E.append(S[l]/NumPar[l])

  plt.plot(NumPar, E, 'b-')
  plt.xlabel('Number of Cores')
  plt.ylabel('Efisiensi')
  plt.show()
```
