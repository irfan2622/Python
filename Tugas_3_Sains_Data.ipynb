```python
from sklearn.ensemble import RandomForestRegressor
# Memecah variabel menjadi fitur dan variabel target 
A = Dataset.drop(['price_range'], axis=1)
B = Dataset['price_range']

# Membuat sebuah  Random Forest regressor dengan 100 pohon
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Sesuaikan model dengan data
rf.fit(A, B)

# Dapatkan  fitur yang penting
importances = rf.feature_importances_

# Membuat sebuah dataframe fitur yaang penting 
feature_importances = pd.DataFrame({'feature': A.columns, 'importance': importances})

# Urutkan dataframe berdasarkan kepentingan dalam urutan menurun
feature_importances = feature_importances.sort_values('importance', ascending=False)

# Print 
print(feature_importances)
```

```python
from sklearn.ensemble import RandomForestRegressor
# Memecah variabel menjadi fitur dan variabel target
A = Datatest.drop(['id'], axis=1)
B = Datatest['id']

# Membuat sebuah  Random Forest regressor dengan 100 pohon
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Sesuaikan model dengan data
rf.fit(A, B)

# Dapatkan  fitur yang penting
importances = rf.feature_importances_

# Membuat sebuah dataframe fitur yaang penting 
feature_importances = pd.DataFrame({'feature': A.columns, 'importance': importances})

# Urutkan dataframe berdasarkan kepentingan dalam urutan menurun
feature_importances = feature_importances.sort_values('importance', ascending=False)

# Print 
print(feature_importances)
```

```python
# Mengecek kolom/fitur terpenting berdasarkan importance dan korelasi

# Menghitung matriks korelasi
corr_matrixS = Dataset.drop(['price_range'], axis=1).corr()

# Menentukan nilai  threshold untuk importance dan correlation
importance_threshold = 0.05
correlation_threshold = 0.8

# Mengidentifikasikan kolom berdasarkan importance dan correlation
important_columns = feature_importances[feature_importances['importance'] >= importance_threshold]['feature']
correlated_columns = set()
for i in range(len(corr_matrixS.columns)):
    for j in range(i):
        if corr_matrixS.iloc[i, j] >= correlation_threshold:
            colname = corr_matrixS.columns[i]
            correlated_columns.add(colname)

# Hapus kolom yang tidak penting atau tinggi korelasi dengan fitur lain 
columns_to_drop = set(A.columns) - set(important_columns) - correlated_columns
A = A.drop(columns_to_drop, axis=1)

# Print 
print(A.columns)
```

```python
f1_score(b_test,b_pred, average='weighted')
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

def evaluation_parametrics(name,b_val, b_pred):
    
    print("\n------------------------{}------------------------\n".format(name))

    cm_test = confusion_matrix(b_val, b_pred)
    t1 = ConfusionMatrixDisplay(cm_test)    
    print("\nClassification Report for Data Test\n")
    print(classification_report(b_val, b_pred))   
    print("--------------------------------------------------------------------------")

    t1.plot()
    
evaluation_parametrics("Machine Learning - Classification", b_test, b_pred)
```
