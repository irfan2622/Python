```python
# Membuat sebuah boolean mask dari  DataFrame untuk mengidentifikasi data yang hilang
missing_values_mask = sale.isnull()

# menghitung data yang hilang
missing_values_count = missing_values_mask.sum()

# membuat sebuah  DataFrame untuk menunjukkan kolom dengan data yang hilang 
missing_values_sale = pd.DataFrame({'column_name': sale.columns, 'missing_values': missing_values_count})

# memfilter sehingga hanya menunjukkan kolom dengan data yang hilang 
missing_values_sale = missing_values_sale[missing_values_sale['missing_values'] > 0]
#print data frame
print(missing_values_sale)
```

```python
# Menghitung proporsi nilai yang hilang di setiap kolom
missing_values_prop = sale.isnull().mean()

# Memfilter sehingga hanya menampilkan kolom dengan nilai yang hilang
missing_cols = missing_values_prop[missing_values_prop > 0]

# print kolom dengan nilai yang hilang dan proporsi nilai yang hilang
print(missing_cols)
```

```python
# Memperoleh nama kolom yang memiliki missing value
cols_to_drop = missing_values_prop[missing_values_prop > 0.1].index

# Melakukan drop kolom dari data frame 
sale.drop(cols_to_drop, axis=1, inplace=True)

# Melakukan print 
sale
```

```python
sale.drop('Id', axis=1, inplace=True) 
# melakukan drop pada ID
```

```python
# Membuat dictionary untuk memetakan data kategorikal ke nilai numerik
mapping = {
    'A': 0,
    'C': 1,
    'FV': 2,
    'I': 3,
    'RH': 4,
    'RL': 5,
    'RP': 6,
    'RM': 7
}

# menggantikan nilai kategorik dengan nilai numerik 
sale['MSZoning'] = sale['MSZoning'].replace(mapping)

# print the updated DataFrame
print(sale)
```

```python
from sklearn.preprocessing import LabelEncoder

# Membuat dictionary untuk memetakan setiap variabel kategori ke objek LabelEncoder yang sesuai
le_dict = {}

def encode_categorical(sale, var): # Mendefinisikan encode kategorik
    le = LabelEncoder()
    sale[var] = sale[var].astype(str)  
    sale[var] = le.fit_transform(sale[var])
    le_dict[var] = le
    return sale

cat_features = sale.select_dtypes(include=['object'])
for column in cat_features.columns:
    sale=encode_categorical(sale,column)
#sale = encode_categorical(sale, 'Street')
#print(le_dict)  # should print {'Street': LabelEncoder()}
sale
```

```python
#Mendefinisikan IQR untuk menghapus data outlier 
Q1 = data1.quantile(q=.25)
Q3 = data1.quantile(q=.75)
IQR = Q3 - Q1
print(IQR)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X1=data1.drop('SalePrice',axis=1)
y1=data1['SalePrice']
```

```python
#Melakukan split data 
lr = LinearRegression()
L=[]
for i in range(1,6):
    X_train, X_test, y_train, y_test = train_test_split(X1.values, y1.values, test_size=0.1*i, random_state=42)
    lr.fit(X_train,y_train)
    y_pred = lr.predict(X_test)
    L.append(r2_score(y_pred,y_test))
print(L)
```
